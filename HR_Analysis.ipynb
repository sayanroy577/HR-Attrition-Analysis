# Exploratory Data Analysis

```python
#1.How many employees have left vs stayed?
ax=sns.countplot(x='Attrition', data=df)
ax.bar_label(ax.containers[0])
plt.title('Attrition Distribution')
```

# Out of 1473, 237 employee has left the company.

```python
#2.Attrition by marital status.
ax=sns.countplot(data=df, x='MaritalStatus', hue='Attrition')
ax.bar_label(ax.containers[0])
plt.title('Attrition by Marital Status')
```

#Attrition level is high among single employee

```python
#3.Job Satisfaction wise attrition
ax=sns.countplot(data=df, x='JobSatisfaction', hue='Attrition')
ax.bar_label(ax.containers[0])
plt.title('Job Satisfaction wise Attrition')
```

#Those who are not satisfied with the job,are likely to leave the company.

```python
#4.Years at company vs attrition
sns.boxplot(data=df, x='Attrition', y='YearsAtCompany')
plt.title('Years at Company vs Attrition')
```

#Significant portion of employees leave the company relatively early(2-3)years in their tenure.

```python
#5.Does OverTime contribute to attrition?
ax=sns.countplot(data=df, x='OverTime', hue='Attrition')
ax.bar_label(ax.containers[0])
plt.title('OverTime vs Attrition')
```

#Overtime plays a major role in attrition.

```python
#6.Average income of leavers vs stayers
df.groupby('Attrition')['MonthlyIncome'].mean().plot(kind='bar', title='Average Monthly Income by Attrition')
```

#Employees leave due to low income.

```python
#7.Years in current role vs attrition
sns.boxplot(data=df, x='Attrition', y='YearsInCurrentRole')
plt.title('Years in Current Role vs Attrition')
```

#Employees who stays relatively long time into their current role,unlikely to leave.

```python
#8.Attrition across different performance ratings.
ax=sns.countplot(data=df, x='PerformanceRating', hue='Attrition')
ax.bar_label(ax.containers[0])
plt.title('Attrition across Different Performance Ratings')
```

#Higher performers (Rating '4') are less likely to attrite.

```python
#9.How does environment satisfaction relate to attrition?
ax=sns.countplot(data=df, x='EnvironmentSatisfaction', hue='Attrition')
ax.bar_label(ax.containers[0])
plt.title('Environment Satisfaction vs Attrition')
```

#Employees attrite more after dissatisfying with the environment.

```python
#10.Heatmap of numerical column.
df['Attrition_Numeric'] = df['Attrition'].astype('category').cat.codes
numerical_df = df.select_dtypes(include=['number']).drop(columns=['Attrition'], errors='ignore')
correlation_matrix = numerical_df.corr()
plt.figure(figsize=(15,12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap with Attrition (Numeric)')
plt.show()
```

#Corelation Insights

*   Job Involvement, Job Satisfaction, Environment Satisfaction, Relationship Satisfaction, WorkLifeBalance: These columns likely show a negative correlation with Attrition_Numeric. This suggests that employees with higher satisfaction or better work-life balance tend to have lower attrition rates. This is a very common and intuitive finding in HR analytics.

*   NumCompaniesWorked: This can also show a positive correlation, indicating that employees who have worked for more companies in the past might be more prone to changing jobs again.
*   YearsAtCompany, YearsInCurrentRole, YearsSinceLastPromotion, YearsWithCurrManager: These usually exhibit negative correlations. Employees who have spent more years at the company, in their current role, since their last promotion, or with their current manager, tend to have a lower likelihood of attrition, indicating stability.
